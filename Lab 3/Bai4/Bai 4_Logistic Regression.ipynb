```python
import numpy as np
import pandas as pd
from google.colab import files
import io
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_absolute_error, mean_squared_error, r2_score
from sklearn import preprocessing
import matplotlib.pyplot as plt
import copy, math
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('students.csv')
df
```

```python
df = df.dropna()
df.shape
```

```python
df.info()
```

```python
df['Target'].unique()
```

```python
df['Target'] = df['Target'].replace('Dropout', 'non-Graduate')
df['Target'] = df['Target'].replace('Enrolled', 'non-Graduate')
df['Target'].unique()
```

```python
encoder = LabelEncoder()
encoder.fit(df['Target'])
df['Target'] = encoder.transform(df['Target'])
df['Target'].unique()
```

```python
dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))
```

```python
X = df.iloc[:,:-1].values
print(f'X shape: {X.shape}, type: {type(X)} \n')
X
```

```python
y = df['Target'].values
print(f'X shape: {y.shape}, type: {type(y)} \n')
y
```

```python
scaler = StandardScaler()

scaler.fit(X)
X_scaled = scaler.transform(X)
print(f'X shape: {X.shape}, type: {type(X)} \n')
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)
```

```python
logistic_reg_model = LogisticRegression()
logistic_reg_model.fit(X_train, y_train)

weights = logistic_reg_model.coef_
bias = logistic_reg_model.intercept_
print(f'best weights: {weights} \nbest bias: {bias}')
```

```python
y_pred = logistic_reg_model.predict(X_test)

for i in range(100):
  print(f'Label: {y_test[i]:<10} | Prediction: {y_pred[i]:<10}')
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
